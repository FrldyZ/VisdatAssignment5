```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

[Peta Jaringan Jawa Barat](http://dbmtr.jabarprov.go.id/wp-content/uploads/2018/11/Peta-Jaringan-Jalan-Prov.-Jawa-Barat-resized.jpg)

```python
df = pd.DataFrame({
    'Big city': [
        'Sukabumi', 'Sukabumi', 'Sukabumi', 'Sukabumi',
        'Bandung', 'Bandung', 'Bandung', 'Bandung', 'Bandung', 'Bandung',
        'Bekasi', 'Bekasi', 'Bekasi', 'Bekasi', 'Bekasi',
        'Bogor', 'Bogor', 'Bogor', 'Bogor', 
        'Cirebon', 'Cirebon', 'Cirebon', 'Cirebon','Cirebon',
        'Tasikmalaya', 'Tasikmalaya', 
        'Karawang','Karawang','Karawang', 'Karawang', 
        'Ciamis','Ciamis','Ciamis','Ciamis',
        'Sumedang','Sumedang','Sumedang','Sumedang',
        'Purwakarta','Purwakarta','Purwakarta','Purwakarta','Purwakarta'
    ],
    'Neighbour': [
        'Bogor', 'Bekasi', 'Bandung',  'Cirebon',
        'Bogor', 'Bekasi', 'Sukabumi', 'Cirebon', 'Tasikmalaya','Indramayu',
        'Bogor', 'Bandung', 'Cirebon', 'Sukabumi','Karawang',
        'Bekasi', 'Sukabumi', 'Bandung', 'Karawang',
        'Bandung', 'Bogor', 'Sukabumi', 'Tasikmalaya','Indramayu',
        'Bandung', 'Cirebon',
        'Bekasi','Bogor','Purwakarta','Bandung',
        'Tasikmalaya','Cirebon','Sumedang','Bandung',
        'Indramayu','Bandung','Cirebon','Tasikmalaya',
        'Karawang','Bekasi','Bandung','Sukabumi','Bandung'
        ]

})

G = nx.from_pandas_edgelist(df,'Big city', 'Neighbour')
```

```python
plt.figure(figsize=(20, 18))
nx.draw_shell(G, with_labels=True, node_size = 20000, node_color=range(11), cmap='gist_rainbow_r')
plt.title("Jalur transportasi yang menghubungkan kota-kota besar di Jawa Barat", fontsize=20);
```

```python
betweenness = nx.betweenness_centrality(G)
degree = nx.degree_centrality(G)

column = {'Betweenness Centrality' : pd.Series(betweenness.values(), index = betweenness.keys()),
          'Degree Centrality' : pd.Series(degree.values(), index = degree.keys())}

df = pd.DataFrame(column)
df.sort_values(by=['Betweenness Centrality', 'Degree Centrality'], ascending=False)
```

**KESIMPULAN**

Berdasarkan nilai tertinggi Betweenness Centrality yaitu adalah kota bandung, dan berdasarkan nilai tertinggi Degree Centality yaitu kota Cirebon,Sehingga bisa disimpulkan bahwa kota terpenting adalah kota bandung diikuti kota cirebon kemudian bekasi dst.
